##7.데이터 전처리하기


###(1)열 정리하기

```python
#시도->광역지방자치 열이름 변경하기
#변수명.rename(columns = {‘열이름’: ‘새로운 열이름’})
accData.rename(columns = {'시도':'광역지방자치'}, inplace= True)
accData
```

```python
#변경 내용 재확인을 위해서 상위 5개 행 확인
#상위 코드에서 열이름을 변경하였지만 실제 데이터에는 반영되지 않음
accData.head()
```

```python
#시도->광역지방자치
#시군구->기초지방자치
#열이름 변경하고 현재 테이블에 결과 반영하기
#변수명.rename(columns = {'열이름':'새로운 열이름'}, inplace= True)
#inplace=True 사용시 이전 열이름 더이상 사용할 수 없음
accData.rename(columns = {'시도':'광역지방자치', '시군구':'기초지방자치'}, inplace= True)

accData.head()
```

```python
#'사고건수' 열 삭제하기
#변수명.drop(columns=['열이름'], axis=1)
accData.drop(columns=['사고건수'], axis=1)
```

```python
#변경 내용 재확인을 위해서 상위 5개 행 확인
#상위 코드에서 열을 삭제하였지만 실제 데이터에는 반영되지 않음
accData.head()
```

```python
#삭제한 열을 원본 테이터에 반영하기
#변수명.drop(columns=['열이름'], axis=1, inplace= True)
#inplace= True 사용 후 '사고건수'열을 더 이상 사용할 수 없음
accData.drop(columns=['사고건수'], axis=1, inplace= True)
```

###(2)행 정리하기

```python
#0번째 행 삭제 후 상위 5개 데이터의 index 확인하기
#변수명.drop(index=행번호, axis=0)
accData.drop(index=0, axis=0)
```

```python
#index 1, 3, 5번째 행 삭제 후 상위 5개 데이터의 index 확인하기
#변수명.drop(index=행번호, axis=0)
accData.drop(index=[1,3,5], axis=0)
```

```python
#index : 0번째~4번째 행 삭제 후 상위 5개 데이터의 index 확인하기
#변수명.index[0:5] 속성 추가
accData.drop(index=accData.index[0:5], axis=0)
```

```python
#index : 0번째~15번째 행 삭제하고 원본데이터에 반영 후 상위 20개 확인하기
#inplce =True 속성 추가
accData.drop(index=accData.index[0:15], axis=0, inplace=True)
accData.head(20)

```

```python
#index 번호 재설정하기
#변수명.reset_index()
accData.reset_index()
```

```python
#기존 index 버리고 새로운 인덱스 설정하며 원본 데이터에 반영하기
#변수명.reset_index(drop=True, inplace=True)
#drop=True : 기존 index 열 삭제됨
accData.reset_index(drop=True, inplace=True)
accData.head()
```
